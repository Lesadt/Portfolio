```python
# 1-Drop last row
df = df.drop(index = 4076)
# 2-First Name and Surname can be put together as 1 column
df['Name'] = df['First name'] + ' ' + df['Surname']
df.insert(0, 'Name', df.pop('Name'))
# 3- Club column is not relevant to the project
df = df.drop(['Club', 'First name', 'Surname'] , axis = 'columns')
df.head()
```

```python
# List of unique countries in serch for errors.
print(df['Country'].unique())
print(len(df['Country'].unique()))

# Detected: Countries in letters code, Words with 1 extra space, 1 'nan'. (Chance rest Manualy)
# Country 3letters code
url = 'https://raw.githubusercontent.com/Lesadt/Portfolio/refs/heads/main/Noseman/Country_Code.csv'
codes = pd.read_csv(url)
country_name = codes['name'].to_list()
country_code = codes['alpha-3'].to_list()
country_codes = dict(zip(country_code, country_name))
```

```python
# 4: Missing Countries
df['Country'].fillna('NONE', inplace = True)

# Extra space after de country name (" ")
for i in range(len(df['Country'])):
    country = df['Country'][i]
    if country[-1] == ' ':
        replacement = country[:-1]
        df['Country'][i] = replacement

# Short Names for Countries (Dictionaire in country_codes)
for i in country_code:
    df['Country'].replace(i, country_codes[i], inplace = True)

# Rest of Countries Manually
faltantes = ['England','Great Britain', 'Scotland', 'California', 'Northern Ireland','UK', 'Belgium & Italy', 'Austrailia', 'Czech Republic','NORWAY', 'Russian Federation', 'Korea, Republic of', 'United States', 'SUI', 'DEN', 'GER', 'RSA', 'NED', 'URU', 'NONE', 'CHI', 'PHI', 'GRE', 'LAT', 'POR', 'TPE', 'INA', 'BRU', 'HON', 'SLO', 'CRO', 'ICE', 'SRI', 'IRE']
fixed_names = ['United Kingdom', 'United Kingdom', 'United Kingdom', 'United States of America', 'United Kingdom', 'United Kingdom', 'Belgium', 'Australia', 'Czechia', 'Norway', 'Russia', 'South Korea', 'United States of America', 'Switzerland', 'Denmark', 'Germany', 'Russia', 'Netherlands', 'Uruguay', 'NONE', 'China', 'Philippines', 'Greece', 'Latvia', 'Portugal', 'Taiwan', 'Indonesia', 'Brunei Darussalam', 'Honduras', 'Slovenia', 'Croatia', 'Iceland', 'Sri Lanka', 'Ireland']

for i in range(len(faltantes)):
    df['Country'].replace(faltantes[i], fixed_names[i], inplace = True)
```

```python
# 5 Chance types and fix
# Swiming half distance in 2015
# Bike was 200km in 2007 and 2011, the rest is 180km
df['Bike distance'] = 180
df.loc[df['Year'] == 2007, 'Bike distance'] = 200 
df.loc[df['Year'] == 2011, 'Bike distance'] = 200 

# Chance all running to 42
df['Run distance'] = 42

# Columns to int:
df = df.astype({'Year':int, 'Swim distance':int, 'Bike distance':int, 'Run distance':int})
```

```python
'''
From here I start with time tables and the re-check for accuracy on T-shirts and finish_lines
Detected errors:
- Missing values on partial times: Inittialy set to '00:00:00'
- Wrong format in T1 and T2 (minutes set like hours)
- Sum does not match total time
''' 
```

```python
# separate times in a different df for testings and changes.
times = df[['Swim time', 'Time T1', 'Cycle time', 'Time T2', 'Run time', 'Total time']]
times = times.fillna('00:00:00')
times['Cycle time'][232] = '08:55:34'
```

```python
#Adding extra 0 at begining so all times have the same structure
for col in times:
    for i in range(len(times)):
        if times[col][i][1] == ':':
            t = '0' + df[col][i]
            times[col][i]  = t
```

```python
#Fixing wrong possition on time values (some minutes are set as hours).
fix_col_T1_T2 = ['Time T1', 'Time T2']
for col in fix_col_T1_T2:
    for i in range(len(times)):
        if str(times[col][i]).count(':') != 2:
            t = '00:' + df[col][i]
            times[col][i]  = t
        elif int(times[col][i][:2]) > 0:
            t = '00:' + times[col][i][:5]
            times[col][i]  = t
```

```python
#Converting times to seconds to have the in 'int' type for easier calculations
def time_to_sec(time_str):
    time = time_str.split(':')
    time = (int(time[0])*3600)+(int(time[1])*60)+(int(time[2]))
    return time

def time_to_str(time_sec):
    hours = str(time_sec // 3600)
    if len(hours) == 1:
        hours = '0' + hours
    minutes = str((time_sec % 3600) // 60)
    if len(minutes) == 1:
        minutes = '0' + minutes    
    seconds = str((time_sec % 3600) % 60)
    if len(seconds) == 1:
        seconds = '0' + seconds    
    time = hours + ':' + minutes + ':' + seconds
    return time
```

```python
# creating a 'times in seconds' table.
time_columns = ['Swim time', 'Time T1', 'Cycle time', 'Time T2', 'Run time', 'Total time']
for i in range(len(times)):
    for col in time_columns:
        to_sec = time_to_sec(times[col][i])
        times[col][i] = to_sec
```

```python
times['Sum_of_Partials'] = times['Swim time']+times['Time T1']+times['Cycle time']+times['Time T2']+times['Run time']
times.head()
diferent = []
for i in range(len(times)):
    if (times['Total time'][i] - times['Sum_of_Partials'][i]) != 0:
        diferent.append(times.loc[i,].values.tolist())
len(diferent)
#Some total times are from 1 to 4 seconds bigger because there where miliseconds not considered in the partial times.
```

```python
for i in range(len(times)):
    diference = times['Total time'][i] - times['Sum_of_Partials'][i]
    if diference == 1:
        times['Run time'][i] += 1
    elif diference == 2:
        times['Run time'][i] += 1
        times['Time T2'][i] += 1
    elif diference == 3:
        times['Run time'][i] += 1
        times['Time T2'][i] += 1
        times['Cycle time'][i] += 1
    elif diference == 4:
        times['Run time'][i] += 1
        times['Time T2'][i] += 1
        times['Cycle time'][i] += 1
        times['Time T1'][i] += 1
```

```python
for i in range(len(times)):
    diference = times['Sum_of_Partials'][i] - times['Total time'][i]
    if diference == 1:
        times['Run time'][i] -= 1
    elif diference == 2:
        times['Run time'][i] -= 1
        times['Time T2'][i] -= 1
    elif diference == 3:
        times['Run time'][i] -= 1
        times['Time T2'][i] -= 1
        times['Cycle time'][i] -= 1
    elif diference == 4:
        times['Run time'][i] -= 1
        times['Time T2'][i] -= 1
        times['Cycle time'][i] -= 1
        times['Time T1'][i] -= 1
```
